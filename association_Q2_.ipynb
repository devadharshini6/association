```python
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
movie = pd.read_csv("/content/my_movies.csv")
movie
```

```python
movie1=movie.iloc[:,5:]
movie1
```

```python
m_itemsets = apriori(movie1, min_support=0.1, use_colnames=True)
m_itemsets
```

```python
c_rules = association_rules(m_itemsets, metric="confidence", min_threshold=0.6)
c_rules
```

```python
l_rules = association_rules(m_itemsets, metric="lift", min_threshold=0.8)
l_rules
```

```python
l_rules.sort_values('lift',ascending = False)[0:20]
```

```python
l_rules[l_rules.lift>1]
```

```python
l_rules[['support','confidence']].hist()
```

```python
l_rules[['support','confidence','lift']].hist()
```

```python
plt.scatter(l_rules['support'], l_rules['confidence'])
plt.show()
```

```python
plt.scatter(c_rules['support'], c_rules['confidence'])
plt.show()
```
